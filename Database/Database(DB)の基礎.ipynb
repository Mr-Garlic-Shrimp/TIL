## ■ Databaseとは何か。なぜ必要か。
あるデータA（例えば、顧客IDごとの購入履歴など）を使った処理を行うアプリが複数あり、  
各アプリごとにデータAを保持・更新するような場合（多重管理の場合）を考える。（大昔のシステムはこうだったらしい）  
この場合、あるアプリがデータAを更新しても別のアプリが持つデータAには反映されず、データの不整合が生じる。  
またデータAにアクセスする仕組みは各アプリで共通部分が多いはずだが、これも各アプリで実装しなければならず、開発の無駄が大きい。  
こうした問題が起きないようにデータを一元管理し、複数のアプリから同時に利用・操作できるようにする仕組みが  
 __Data Base Management System(DBMS)__ であり、DBMSに管理されたデータの集まり自体を __Database(DB)__ という。  

## ■ DBMSの基本機能
* __永続的(persistent)記憶__  
アプリの実行／終了に関わりなく、データの内容を保ちつづける。
* __データ操作__  
データベースのテーブル（表）に対する、挿入・更新・削除などの操作を実行する。
* __クエリ（問合せ）__  
データの内容に関するアプリからの問合せを受け付け、データベースにアクセスして応答する。  
問合せには、SQLというプログラム言語（クエリランゲージ）を使う。
* __2次記憶管理__  
データを格納しているデータベースファイルを管理する。
* __障害回復__  
ハードウェア障害やソフトウェアのエラー（バグ）が発生し、データ内容が壊れたとき、  
トラブル発生前のデータを復元できる。
* __並列処理制御（トランザクション制御）__  
複数のアプリから、同じデータベースに同時にアクセスした場合にも、DB内のデータに不整合が生じないようにする。
