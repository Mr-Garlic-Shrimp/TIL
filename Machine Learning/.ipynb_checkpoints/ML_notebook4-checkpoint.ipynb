```python
%load_ext autoreload
%autoreload 2
import pandas as pd
import numpy as np
import seaborn as sns
from itertools import product
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
import statsmodels.api as sma
from sklearn.model_selection import train_test_split ,LeaveOneOut, cross_val_score, KFold, RepeatedKFold
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsRegressor

%matplotlib inline
import matplotlib.pyplot as plt
```

## ○ Ridge

```python
# サンプルデータを取得。pd.read_csvはhttpsでも取ってこれる。
# メジャーリーグの選手の成績と給料等のデータ
df = pd.read_csv('https://raw.githubusercontent.com/kirenz/datasets/master/Hitters.csv')
```

```python
# Salaryはの単位は1,000$
df
```

```python
# 統計量を見てみる
df.describe()
```

```python
# Salaryのcountが他とあっていない　→NaNがあるのでdropnaする
df.dropna(inplace=True)
```

```python
df.describe()
```

```python
# データ準備
y_col = 'Salary'
# 下記のような書き方でも目的変数以外のカラムを取り出せる
X = df.loc[:, df.columns!=y_col]
y = df[y_col]

# 標準化のために量的変数のカラム名のリストを作成。これは質的変数をダミー変数に変換する前にやること。
numeric_cols = X.select_dtypes(include=np.number).columns.to_list()

# ダミー変数作成
X = pd.get_dummies(X, drop_first=True)

```

```python
numeric_cols
```

```python
X.head()
```

```python
# hold-out
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
# 量的変数のみ標準化(学習データ、テストデータ両方)
scaler = StandardScaler()
X_train[numeric_cols] = scaler.fit_transform(X_train[numeric_cols])
X_test[numeric_cols] = scaler.transform(X_test[numeric_cols])
```

```python
X_train.head()
```

```python
X_train.describe()
```

```python
X_test.describe()
```

X_testを標準化しても平均0、標準偏差は1にならないことに注意。  
（学習データの平均と標準偏差を使って計算しているため。）

```python
# Ridgeでモデル学習。λのデフォルト値は1.0
ridge = Ridge()
ridge.fit(X_train, y_train)
```

```python
# Salaryを予測
y_pred = ridge.predict(X_test)
y_pred
```

```python
y_test
```

```python
# y_predとy_testの残差を算出
mse = mean_squared_error(y_test, y_pred)
mse
```

```python
# MAEを計算
mean_absolute_error(y_test, y_pred)
```

平均して251Kドル(3000万円ぐらい)、真の値からずれている。

```python
# パラメタを見てみる
ridge.coef_
```

```python
# 単純な線形回帰の場合も見てみる
lr = LinearRegression()
lr.fit(X_train, y_train)
lr.coef_
```

```python
y_pred_lr = lr.predict(X_test)
mean_absolute_error(y_test, y_pred_lr)
```

単純な線形回帰の場合とRidgeを比べると、Ridgeの方がパラメタの絶対値が小さくMAEも小さい（精度が良い）ことが分かる。

## ○ λを変えたときのRidgeの精度を比較する
さまざまな特徴量からSalaryを予測するモデルを作成する。  
* $\lambda=10^{-3}～10^{3}$（間隔は任意）
* hold-outで学習データとテストデータを7:3で分割
* 評価指標にはMSEを採用
* $\lambda$を増加させたときのMSEの推移をplot
* $\lambda$を増加させたときの特徴量の係数$\theta_j$の推移をplot

```python
# λの用意
alpha_ridge = np.logspace(-3, 3, 50)

# MSEとパラメタを格納するリストを定義
result_mse = []
result_coef = []

# hold-out
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# 量的変数のみ標準化(学習データ、テストデータ両方)
scaler = StandardScaler()
X_train[numeric_cols] = scaler.fit_transform(X_train[numeric_cols])
X_test[numeric_cols] = scaler.transform(X_test[numeric_cols])


# Ridgeでモデル学習。
for alpha in alpha_ridge:    
    ridge = Ridge(alpha)
    ridge.fit(X_train, y_train)
    y_pred = ridge.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    result_mse.append(mse)
    result_coef.append(ridge.coef_)
    #result[f'{i}'] = {'lambda of Ridge':alpha, 'MSE': mse, 'coef': ridge.coef_}
    
```

```python
# λとMSEをプロット
plt.plot(alpha_ridge, result_mse)

# 横軸を対数表記にする
plt.xscale('log')
plt.xlabel('lambda of Ridge')
plt.ylabel('MSE')
```

λを極端に大きくすると、パラメタが小さくなる＝モデルが単純化され、biasが大きくなってしまうので精度が下がる。

```python
# 任意の特徴量のパラメタの推移をリストで格納する
coef_i = 0
coef_hist = [coef[coef_i] for coef in result_coef]
```

```python
# λとcoefの推移をプロット
plt.plot(alpha_ridge, np.abs(coef_hist))

plt.xscale('log')
plt.xlabel('lambda of Ridge')
plt.ylabel('coef')
```

教訓：下手に見やすいように辞書型で出力とかするよりリストやnumpyで出力した方が扱いやすいかも。

```python
# 全てのパラメタの推移を見てみる。
plt.plot(alpha_ridge, np.abs(result_coef))

plt.xscale('log')
plt.xlabel('lambda of Ridge')
plt.ylabel('coef')
```

全ての係数が0に近づいていっていることが分かる。  
Ridgeではある係数だけ途中でゼロになるなどはなく、全ての係数で同じようなタイミング（λ）で徐々に0に近づく。

## ○ Lasso
sklearn.linear_model.Lassoクラスを使う
1. Lasso(alpha)でインスタンスを生成(alphaには$\lambda$を入れる)
2. .fit(X, y)で学習
3. .predict(X)で予測

```python
# デフォルトではλは1.0
lasso = Lasso() 
lasso.fit(X_train, y_train)
y_pred = lasso.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred) 
mse
```

```python
mae = mean_absolute_error(y_test, y_pred)
mae
```

```python
lasso.coef_
```

Ridgeと違い、一部の係数が完全に０になっていることが分かる。  
すなわち、０になった係数が係る特徴量は目的変数に寄与しないので分析対象外としてよい。
