```python
%load_ext autoreload
%autoreload 2
import pandas as pd
import numpy as np
import seaborn as sns

%matplotlib inline
import matplotlib.pyplot as plt
```

## ○シグモイド関数を描く

```python
fig, ax = plt.subplots(figsize=(10,5))
x = np.linspace(-10,10, 100)
y = 1 / (1 + np.exp(-x) )
ax.plot(x,y)
# r'$Latex表記$'で軸名やタイトルにLatex表記が使える。
ax.set(title='Sigmoid', xlabel=r'$x$', ylabel='z')
```

## ○LogLossを描画

```python
fig, ax = plt.subplots(figsize=(10,8), nrows=2)
p = np.linspace(0,1, 100)
y1 = -np.log(p)
y2 = -np.log(1-p)
ax[0].plot(p,y1)
ax[1].plot(p,y2)
# r'$Latex表記$'で軸名やタイトルにLatex表記が使える。
ax[0].set(title=r'$y_i=1$', xlabel=r'$p(x_i)$', ylabel=r'$-log(p(x_i))$')
ax[1].set(title=r'$y_i=0$', xlabel=r'$p(x_i)$', ylabel=r'$-log(1-p(x_i))$')
ax[0].set_ylabel(r'$-log(p(x_i))$',rotation=0,labelpad=30)
ax[1].set_ylabel(r'$-log(1-p(x_i))$',rotation=0,labelpad=35)


fig.tight_layout()
```
