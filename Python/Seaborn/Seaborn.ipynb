# Seabornメモ
Matplotlibよりもスタイリッシュなグラフを描くためのライブラリ。プレゼンなどに使えるので業務でもよく使う。  
matplotlibのラッパーライブラリ（内部でmatplotlibを使っている）である。

```python
import seaborn as sns
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## ヒストグラム：sns.distplot() ※現在はdisplotが推奨

```python
df = pd.read_csv('../pandas/sample_data/tmdb_5000_movies.csv')

# 縦軸は０から１で正規化された状態（割合）で表示される。
sns.distplot(df['vote_count'])
```

実線は確率密度関数（probablility density function: PDF）。  
正規化されているので、全体の面積は1。この場合、ある範囲の面積は確率として扱える。  
確率変数（ここでは横軸の値）がある値の範囲に入っている確率を求めるには  
その範囲で確率密度関数を横軸の変分で積分して面積を求めればよい。  
カーネル密度推定で確率密度関数を求めている。


```python
# 正規化とカーネル密度推定をオフにする場合は下記引数をFalseにする。
sns.distplot(df['vote_count'], norm_hist=False, kde=False)
```

## 散布図

### sns.joint()

```python
# 各軸のヒストグラムも表示してくれる
sns.jointplot(x='vote_count', y='vote_average', data=df)
```

```python
df = pd.read_csv('../pandas/sample_data/tmdb_5000_movies.csv')
# vote_countが少ないものは信憑性がないので除く
df = df[df['vote_count'] >= 100]

# kind=regとすると、変数間の関係性をざっくり曲線で表現する。
sns.jointplot(x='vote_count', y='vote_average', data=df, kind='reg')
```

```python
# kind='kda'とすると、縦軸、横軸の確率密度関数を２次元で表示する。山を上から見ているイメージ
sns.jointplot(x='vote_count', y='vote_average', data=df, kind='kde')
```

### sns.pairplot()
現場でよく使う

```python
# DataFrame内の数値項目の全組み合わせに対してグラフを出力する。自身のカラム同士の個所はヒストグラムが表示される。
sns.pairplot(df)
```

```python
# 必要なカラムのみでフィルタリング
sns.pairplot(df[['revenue', 'budget', 'vote_count', 'vote_average']])
```

```python
# vote_averageが下位50％だったらlow,上位50％だったらhigh
# 速度を求めるならばmedian()は別変数にした方が良い
df['evaluation'] = df['vote_average'].apply(lambda x: 'high' if x > df['vote_average'].median() else 'low')
# hueに判定用の列を指定することで、値ごとに色分けしてくれる。この場合のalphaは下記のように指定する。
sns.pairplot(df[['revenue', 'budget', 'vote_count', 'vote_average', 'evaluation']],
             hue='evaluation',
             plot_kws={'alpha':0.3})
```

## Categorical Plot
<重要ポイント>Seabornのメソッドの戻り値はaxisオブジェクトである。

### sns.barplot()
カテゴリごとに統計量を計算したうえで、棒グラフを描画してくれる。

```python
#　横軸はカテゴリ
# デフォルトでは縦軸の平均値がプロットされる
# 縦の黒い棒はエラーバー。95%の信頼区間（confidence interval)。
# 母集団（ここでは全世界のすべてのデータ）のhigh&lowの値を標本（読み込んだデータ）から推測するときに信頼区間は使う。
# この信頼区間はざっくり言って、母集団から100回標本を取り出して平均値を求めたときにその95回はこの区間に母集団の平均が入っているだろうということを示している。
sns.barplot(x='evaluation', y='revenue', data=df)
```

```python
# 平均値で見た時よりも差が小さい→アバターなど圧倒的な値の影響がないため
sns.barplot(x='evaluation', y='revenue', data=df, estimator=np.median)
```

### sns.countplot()

```python
df['year'] = df['release_date'].apply(lambda x: x[:4])
```

```python
# matplotlibのaxesオブジェクトが返ってくるので、axesに対する操作が可能
ax = sns.countplot('year', data=df, hue='evaluation')

# ラベル（ticks）の向きとサイズを変える
ax.tick_params(rotation=90, labelsize=6)
```

```python
# 年代カラム(decade) 2012 -> 2010s
df['decade'] = df['year'].apply(lambda x: x[:3] + '0s')
df.head()
```

```python
# 横軸がきれいに並ぶようにsort_valuesを使用
sns.countplot(x='decade', data=df.sort_values('decade'), hue='evaluation')
```

### sns.boxplot()

```python
sns.boxplot(x='decade', y='vote_average', data=df.sort_values('decade'))
```

### sns.violinplot()

```python
# どのあたりにデータが集中しているのか、太さで表現する。
sns.violinplot(x='decade', y='vote_average', data=df.sort_values('decade'))
```

### sns.swarmplot()

```python
# swarmplotではさらに、どのあたりにデータが集中しているのか実際のデータ数をプロットして表現する。
sns.swarmplot(x='decade', y='vote_average', data=df.sort_values('decade'))
```

```python
# 1980s以降もプロットが見やすいようにグラフの幅を変える
fig, ax = plt.subplots(figsize=(25, 5))
ax.tick_params(labelsize=30)
# swarmplotのaxというパラメータにsubplotsの戻り値のaxを入れる
sns.swarmplot(ax=ax, x='decade', y='vote_average', data=df.sort_values('decade'))
```

boxplotとswarmplotを並べてプロットすると、パーセンタイル値とデータの集中具合がわかりやすくなる。

## Heatmap
表の各値の大きさに応じて色付けしたもの。  
データの傾向と値の大小を一目で把握するのに使える。  
データサイエンスでは特に相関表や混同行列によく使う。  
* 相関表：相関係数（Rの二乗値）を表にしたもの
* 混同行列：分類機（AIモデル）が上手く分類できたもの・できなかったものを表にしたもの。  
参考：https://datawokagaku.com/seaborn_heatmap/

### .corr()

```python
df = pd.read_csv('../pandas/sample_data/tmdb_5000_movies.csv')
# 相関係数を表示。corr:corerationの略
df.corr()
```

相関係数は-1から1の値をとる。1に近いと正の相関があるといい、一方が増えるともう一方も増える関係を表す。  
負の相関はその逆。

```python
# 信頼性が低い値のチェック。minがゼロは明らかにおかしいので抜く。
df.describe()
```

```python
# 全てのカラムがTrue（0でない）の行のみ抽出。
(df != 0).all(axis=1)
```

```python
# この場合、locでTrueの行のみ表示される。locの引数は行名を指定することが多いが、booleanの配列でもよい。
# df[(df != 0).all(axis=1)]でもOK
df.loc[(df != 0).all(axis=1)]
```

```python
# ゼロを抜いた結果で再出力
df = df.loc[(df != 0).all(axis=1)]
df.corr()
```

```python
## heatmapを作成.数値を入れたいときはannot=Trueとする
fig, ax = plt.subplots(figsize=(10, 5)) 
sns.heatmap(df.corr(), ax=ax,annot=True, cmap='coolwarm')
```

```python
df['year'] = df['release_date'].apply(lambda x: x[:4])
df['month'] = df['release_date'].apply(lambda x: x[5:7])
df.head()
```

```python
# 右のように書くと、単なるrevenueのシリーズとして返ってきてしまう。df2 = df.groupby(['year','month']).mean()['revenue']
# ピボットテーブルとして使うためにDataFrameを受け取るためには[['revenue']]のようにリストで渡す。
# ピボットテーブルは下記のように書く
revenue_df = df.groupby(['year','month']).mean()[['revenue']].pivot_table(values='revenue', index='month', columns='year')
revenue_df
```

```python
sns.heatmap(revenue_df)
```

図の濃淡はrevenueの大きさを表わす。  
5-7月、11-12月はrevenueが高くなる傾向と言えそう。（クリスマスなど）→映画にはシーズンがある。
    

## Seabornのスタイル変更

```python
# context: 全体のスタイルを変更。'paper','notebook','talk','poster'がある
sns.set(context='paper')
sns.displot(df['vote_average'])

# コンテキスト設定用の関数でも設定可能
sns.set_context('notebook')
```

```python
# style: 'darkgrid','whitegird','dark','white','ticks'　→グラフ描画部分のスタイルを変更する。
# ticksは余分な枠が消えてすっきりするのでお勧め
sns.set(context='notebook', style='ticks')
sns.displot(df['vote_average'])
# 下記で軸を消せる。
# sns.despine(bottom=True)
```

```python
df = pd.read_csv('../pandas/sample_data/tmdb_5000_movies.csv')
df.head()
```

```python
# decadeのカラムを作る（復習）
# nanを抜く。df['release_date'].isna()でフィルタリングしてもOK
df = df.dropna(subset=['release_date'])
df['decade'] = df['release_date'].apply(lambda x: x[:3] + '0s')

# vote_averageのmedianでhigh,lowを分ける(復習)
df['evaluation'] = df['vote_average'].apply(lambda x: 'high' if x > df['vote_average'].median() else 'low')  
df.head()
```

```python
# palette引数： deep,muted,bright,pastel,dark,colorblind →hueの色分けを変えるだけ。
# 他のスタイルは'matplotlib colormap'で検索
sns.set(context='notebook', style='ticks', palette='bright')
# plot自体の色はこっちのpaletteで指定する。
sns.violinplot('decade', 'popularity', data=df.sort_values('decade') ,hue='evaluation', palette='twilight')
```

```python
# 付属情報のつけ方。→matplotlibを中で呼んでいるので、同じように設定可能
sns.distplot(df['vote_average'])
plt.title('Distribution of scores')
plt.xlabel('Evaluation Scores')
plt.savefig('saved_seaborn.png')
```

```python
# オブジェクト指向での扱い方
fig, axes = plt.subplots(1, 2)
sns.distplot(df['vote_average'], ax=axes[0])
sns.distplot(df['popularity'], ax=axes[1])
```
